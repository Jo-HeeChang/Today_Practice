```python
comments = [ ]
cnt = 0
for i in range(1,10):
    url = url + str(i)
    page = urlopen(url)
    soup = BeautifulSoup(page, "lxml")

    comment_all = soup.find_all('td', class_='title')
    for comment in comment_all:
        temp= list(comment.children)
        if len(temp) < 5:
            cnt= cnt + 1
            print("len<5 case :",cnt)
            continue
        else:
            try:
                cnt= cnt + 1
                result = temp[6].strip()
                comments.append(result)
            except:
                cnt= cnt + 1
                print("len>=5 case ",cnt)
                print(temp)
print(len(comments))
print(comments)
print(cnt)
```

```python
import pandas as pd
```

```python
dict_doc = {'text' : comments}
doc = pd.DataFrame(dict_doc)
```

```python
doc.to_csv('크루엘라리뷰.csv', index = False)
```
