### Scatter plot
  * 언제 사용한다. 두 변수간의 상관관계를 볼 때 사용한다.

```python
df.iplot(kind='scatter', x='A',y='B',mode='lines+markers',size=20)
```

```python
df.iplot(kind='scatter', x='A',y='B',mode='markers', 
        size=20, color='#3f9142', symbol=10)
```

### 실습

```python
df.iplot(kind='line', x='A', y = 'C',mode = 'markers', size = 20,
         color = '#2f1120',
         symbol = 111,
         title = 'A와 C의 산점도 그래프',
         xTitle = 'A',
         yTitle = 'C')
```
